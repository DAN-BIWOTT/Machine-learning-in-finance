```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('telecom_churn.csv')
```

```python
data.columns
```

```python
# I am plotting Grey to represent those with plans and violet to represent those without plans.
Color = data['Voice mail plan'].map({
    'Yes': 'Grey',
    'No': 'Violet'
})
Color.head()
```

```python
fig = plt.figure(figsize=(11,7))
plt.scatter(data['Voice mail plan'], data['Total day minutes'], color = Color, s = data['Total day minutes'] + 0.05 );
plt.xlabel('Voice mail plan');
plt.ylabel('Total day minutes');
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.boxplot(x = 'Voice mail plan', y = 'Total day minutes', data=data, palette = "Set1");
```

```python
sns.countplot(x = 'Total day minutes', hue = 'Voice mail plan', data = data);
```
